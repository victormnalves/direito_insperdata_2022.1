```python
df1['ministro_relator'], df1['ministro_relator_acordao'] = df1['ministro_relator'].str.split('\spara\sAcórdão', 1).str # separando os ministros entre relatores e relatores do acórdão
df1.loc[df1['ministro_relator_acordao'].isnull(),'ministro_relator_acordao'] = df1['ministro_relator'] # caso não haja um relator para o acórdão, este será o relator "original"
```

```python
resultado = []
for decisao in range(0, len(df1.ementa)):
    if re.search(r"(parcial(mente)? provi(do|mento)|provido(s)? em parte)", df1['ementa'][decisao], flags=re.I | re.S):
        resultado.append("provimento parcial")
    elif re.search(r"([Nn][ãa]o |[Dd]es|[Ii]m|[Nn]egar |NÃO |NEGAR )(provido|conhecido|provimento|PROVIDO|PROVIMENTO)(s)?", df1['ementa'][decisao], flags=re.I | re.S):
        resultado.append("não provimento")
    else:
        resultado.append("provimento")
df1['resultado'] = resultado
df1
```
